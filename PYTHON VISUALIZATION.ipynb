```
v1_1DF = spark.table("zip_level_risk_v1_1")
```

```
v2_2DF = spark.table("zip_level_risk_v1_2")
```

```
v2_2DF.show(10)
```

```
v1_1_reducedDF =v1_1DF.select(["zipcode","count_property"])\
.filter(v1_1DF['zipcode']<=1050)\
.withColumnRenamed("count_property","cp_v1_1").withColumnRenamed("zipcode","zcode_v1_1")
```

```
v1_1_reducedDF.show(5)
```

```
v1_2reducedDF =v1_1DF.select(["zipcode","count_property"])\
.filter(v1_1DF['zipcode']<=1050)\
.withColumnRenamed("count_property","cp v1_2").withColumnRenamed("zipcode","zcode_v1_2")
```

```
v1_2reducedDF.show(8)
```

```
v1_jnt = v1_1_reducedDF.join(v1_2reducedDF,v1_1_reducedDF.zcode_v1_1 == v1_2reducedDF.zcode_v1_2,"inner")
```

```
v1_jnt.show(5)
```

```
v1_rdd=v1_jnt.rdd
```

```
v1_rdd.take(5)
```

```
x_axis = v1_rdd.map(lambda b: b[0]).collect()
```

```
type (x_axis)
```

```
cp_v1_1 = v1_rdd.map(lambda b:b[1]).collect()
```

```
cp_v1_2 = v1_rdd.map(lambda b:b[3]).collect()
```

```
versions = ['cp_v1_1','cp_v1_2']
data = {
  'names':x_axis,
   'cp_v1_1':cp_v1_1,
    'cp_v1_2':cp_v1_2
}
```

```
from bokeh.io import output_file,show
from bokeh.plotting import figure
from bokeh.resources import CDN
from bokeh.embed import file_html

output_file("vision.html")

colors = ["#c9d9d3","#718dbf"]

view = figure(x_range = [1000,1050],height=350, title="Property counts by version",
            toolbar_location=None , tools="hover",tooltips = "$name @names:@$name")
view.vbar_stack(versions,x='names',width=0.9,color =colors, source=data,
             legend_label=versions)

view.legend.location = "top_left"
view.legend.orientation = "horizontal"

view = file_html(view,CDN, "vision.html")
```
